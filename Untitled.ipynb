```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
data = sns.load_dataset("fmri")

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Create the line plot
plt.figure(figsize=(12, 6))
line_plot = sns.lineplot(
    data=data,
    x="timepoint",
    y="signal",
    hue="event",
    style="event",
    markers=True,
    dashes=False
)

# Add titles and labels
line_plot.set_title("FMRI Response Over Time", fontsize=20)
line_plot.set_xlabel("Timepoint", fontsize=14)
line_plot.set_ylabel("Signal", fontsize=14)

# Customize legend
line_plot.legend(title="Event", title_fontsize='13', fontsize='11')

# Show the plot
plt.show()
```
